```python
import pandas as pd
data = pd.read_csv("breast-cancer-wisconsin.csv")
# print(data)
# data.describe()
# data.shape
# data['class'].value_counts()

X1 = data.loc[:, 'clump_thickness' : 'mitoses']
# X1.shape
y = data[['class']]
# y.shape

# X_train, X_test, y_train, y_test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X1, y, stratify = y, random_state = 42)

# print(X_train.mean() - X_test.mean())
# print(y_train.mean())
# print(y_test.mean())

#from sklearn.metrics import mean_squared_error

#RMSE = mean_squared_error(y, y_pred)**0.5

from sklearn.preprocessing import MinMaxScaler
scaler_minmax = MinMaxScaler()
scaler_minmax.fit(X_train)
X_scaler_minmax_train = scaler_minmax.transform(X_train)

# print(type(X_scaler_minmax_train))
# pd.DataFrame(X_scaler_minmax_train).describe()

scaler_minmax.fit(X_test)
X_scaler_minmax_test = scaler_minmax.transform(X_test)

# pd.DataFrame(X_scaler_minmax_test).describe()

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_scaler_minmax_train, y_train)

pred_train = model.predict(X_scaler_minmax_train)
model.score(X_scaler_minmax_train, y_train)

pred_test = model.predict(X_scaler_minmax_test)
model.score(X_scaler_minmax_test, y_test)

from sklearn.metrics import confusion_matrix
confusion_train = confusion_matrix(y_train, pred_train)

# print(confusion_train)

prob_train = model.predict_proba(X_scaler_minmax_train)
pred_train = pd.DataFrame(pred_train)

print(prob_train.shape)
print(pred_train.shape)

y_train[['y_pred']] = pred_train
y_train[['y_proba0', 'y_proba1']] = prob_train

y_train
```
